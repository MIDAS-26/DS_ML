```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
#to get a bar plot which uses average data as default use sns.barplot(x='column name',y='column name',data=dataframe)
#since the default estimator function used is average, to use some other cutom function pass it in the argument estimator=func
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
#to get a count plot in which you get a count of categorical data use sns.countplot(x='column name',data=dataframe)
sns.countplot(x='sex', data=tips) #remember to only pass in x axis as the y axis has already been chosen- the count
```

```python
#to get a box plot to show the distribution of categorical data use sns.boxplot(x='column name',y='column name', data=dataframe)
#the boxes show the quartiles of the dataframe while the whiskers show the rest of the distribution, the points outside the whiskers are the outliers
#the whisker from the bottom to the bottom of the box represent 1 quater as in a quater of the points lie there, then bottom of the box to middle of the box 2nd quater, middle to top 3rd quater, top to top whisker 4th quater and the points outside that are outliers
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
#to add a hue to the box plot and get another categorical column data represented in the plot use argument hue='column name'
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
# to plot a violin plot use sns.violinplot(x='column name', y='column name', data=dataframe)
#violin plot is similar to box plot but it offers more info at the cost of less readability, it's harder to interpret
#violin plot shows all the points that are to be plotted unlike box plot
#violin plot too has a hue argument to represent another categorical data column
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
#instead of getting 2 plots to represent 2 categories mentioned in hue, use split=bool argument to represent the categories in 1 violin plot per x axis data itself
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', split=True)
```

```python
#to get a basic scatter plot use sns.stripplot(x='column name', y='column name', data=dataframe)
#hue and split arguments are also supported. split has been renamed to dodge for this plot
sns.stripplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
#since it's hard to see how many points are stacked on top of each other, use jitter=bool argument to add random noise to the data in order to see the points clearly
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='smoker', dodge=True)
```

```python
#to get a swarm plot use sns.swarmplot(x='column name', y='column name', data=dataframe)
#swarm plot is combination of scatter plot and violin plot
#scalability is low and computation power is high, so avoid this for large dataframes
sns.swarmplot(x='day', y='total_bill', data=tips)
```

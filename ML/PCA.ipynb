```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
%matplotlib inline
```

```python
#using scikit learn's inbuilt cancer dataset, which has real world data
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
#the type of scikit learn datasets are unique
type(cancer)
```

```python
cancer.keys()
```

```python
#don't directly call columns as you'll get a weird formatted text, so use print()
print(cancer['DESCR'])
```

```python
#creating dataframe with desired columns of data
df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
df.head()
```

```python
df.info()
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
#transforming to just 2 principal components
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(12,8))
plt.scatter(x_pca[:,0], x_pca[:,1], c=cancer['target'])
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
#every row is a principal component and every column is a feature name in cancer dataset
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_, columns=cancer['feature_names'])
df
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df_comp, cmap='coolwarm')
```

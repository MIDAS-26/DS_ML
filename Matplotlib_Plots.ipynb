```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#for jupyter notebook, use this to get plot in the notebook itself
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

Functional Method

```python
#plt.plot(x axis data, y axis data, extra stuff like color of line)
plt.plot(x, y)
plt.xlabel('Label for x axis')
plt.ylabel('label for y axis')
plt.title('Title for the plot')
```

```python
#to create multiple plot on the same canvas use plt.subplot(no. of rows, no. of columns, plot no.)
#plt.plot(x axis data, y axis data, extra stuff like color of line)
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

Object Oriented Method

```python
#creating a blank canvas using plt.figure()
fig = plt.figure()

#now to add axes to the canvas using fig_name.add_axes(a list of [distance from y, distance from x, width wrt canvas, height wrt canvas] for the figure)
#all the values in the list should be between 0 and 1
axes = fig.add_axes([0.1,0.1,0.8,0.8])

#now to plot use axes_name.plot(x data, y data)
axes.plot(x,y)

#setting labels
axes.set_xlabel('Label for x')
axes.set_ylabel('Label for y')
axes.set_title('Title for plot')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'r')
axes2.plot(y,x,'b')

axes1.set_title('Larger plot')
axes2.set_title('Smaller plot')
```

```python
#to call plt.figure() in a better way and to make our work easier we use plt.subplots(nrows=int, ncols=int)
fig,axes = plt.subplots(nrows=1, ncols=2)
#it's the same as plt.figure() but the subplots function calls figure, axes for you

#the axes object is an array. Hence, iterable. So we use axes[index] to specify the axes to which you wanna plot on
axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[0].set_xlabel('Label for x1')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
axes[1].set_xlabel('Label for x2')
```

```python
fig,axes = plt.subplots(nrows=4, ncols=5)
```

```python
#as seen above, there's overlapping of axes. So we use plt.tight_layout() to remove overlapping and give all axes enough space
fig,axes = plt.subplots(nrows=4, ncols=5)
plt.tight_layout()
```

```python
#to set a custom figure size and dpi use figsize(width,height) as argument which works with both figure() and subplots()
fig,ax = plt.subplots(nrows=2,ncols=1, figsize=(8,2))
ax[0].plot(x,y)
ax[1].plot(y,x)
plt.tight_layout()
```

```python
#to save the figure use fig.savefig('name.extension_you_want', dpi=int(optional. to get a desired detailed image))
fig.savefig('subplots1.png',dpi=200)
```

```python
#plotting multiple plots on the same axes and labelling them using label argument and ax.legend() function
#for location of legend ax.legend(loc=int from 0 to 10, refer docs for meaning of each int) or ax.legend(loc=(x coordinate, ycoordinate))
fig,ax = plt.subplots()
ax.plot(x,x**2, label='X Squared')
ax.plot(x,x**3, label='X Cubed')
ax.legend(loc=(0.5,0.8))
```

Arguments in plot() function

```python
#defining basic and custom colors for the plots using color='basic_colors or rgb hex code'
#defining line width using linewidth=int or in short lw=int
#defining transparency of the line in plot using alpha=int(1 means opaque)
#defining line style using linestyle='desired_style(eg. '--', '-.', ':', 'steps')' or in short ls=''
#to mark the points that are plotted using marker='desired_marker(eg. 'o', 's', '*', '+', '1')'
#to change the size of the marker using markersize=int(1 is smallest) or in short ms=int
#to change marker face color, edge width, edge color using markerfacecolor='', markeredgewidth=int, markeredgecolor=' or in short mfc='', mew=int, mec=''
fig,ax = plt.subplots()
ax.plot(x,y, color='purple', lw=3, alpha=0.4, ls='-', marker='o', ms=10, mfc='yellow', mew=3, mec='red')

#to set limits on the x, y axes to define how much to show using ax.set_xlim([lower_bound, upper_bound])
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```

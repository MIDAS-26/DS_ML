```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df1 = pd.read_csv('df1', index_col=0)
df1
```

```python
df2 = pd.read_csv('df2')
df2
```

When using pandas for plots, it calls matplotlib under the hood, so you can use all the cutomisations of matplotlib like figsize=(), bins, etc.

```python
#to plot using pandas call the dataset or the column you wanna plot then .plot.plot_you_want()
df1['A'].plot.hist()
```

```python
df2.plot.area(alpha=0.5)
```

```python
df2.plot.bar(stacked=True)
```

```python
df1.plot.line(lw=1, figsize=(12,3))
```

```python
df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm') #the c here is to plot more info based on color then using a different color scheme ny using cmap.
```

```python
df1.plot.scatter(x='A', y='B', s=df1['C']*100) #instead of color, plotting more info wrt size of the data wrt column C.
```

```python
df2.plot.box()
```

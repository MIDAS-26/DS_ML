```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips
```

```python
flights
```

```python
#to represent data in matrix plots the data should be in a matrix
#so to convert a dataframe to matrix use dataframe.corr() to get a correlation matrix or use dataframe.pivot_table()
tc = tips.corr()
tc
```

```python
#to get a heatmap use sns.heatmap(matrix)
#to get annotations in the heatmap use argument annot=bool
sns.heatmap(tc, annot=True)
```

```python
#using pivot to make a matrix for flights dataframe
fp = flights.pivot_table(index='month', columns='year', values='passengers')
fp
```

```python
#to set linewidth and linecolor use the words as arguments
sns.heatmap(fp,linecolor='white',lw=1)
```

```python
# to get a clustermap use sns.clustermap(matrix)
#it divides the data into clusters. Hence the months, years are not in order as they are clustered according to similar data months
#to use a different color combonation use cmap='color combination' argument
#to normalise the data use argument stnandard_scale=int
sns.clustermap(fp, cmap='coolwarm', standard_scale=1)
```

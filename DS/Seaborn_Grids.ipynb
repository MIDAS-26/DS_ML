```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
#pair plot does most of the work on it's own, now we'll use PairGrid() to get more control over the plots
#using just PairGrid(), it'll make empty subplots
g = sns.PairGrid(iris)
```

```python
#now to define a plot use g.map(plt.plot_name)
g.map(plt.scatter)
```

```python
#to customise plots on diagonal, upper & lower parts of the subplot use g.map_diag/upper/lower(plot function)
g.map_diag(sns.histplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
#to get plots with even more info like more categories use sns.FacetGrid(data=dataset, col='col name', row='col name')
g = sns.FacetGrid(tips, col='time',row='smoker') #this gives us empty grids, then plot the plots on these
g.map(sns.histplot, 'total_bill') #now this plots the info based on the col, row defined in the facetgrid function
```
